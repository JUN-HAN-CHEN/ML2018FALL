```python
#Allowable Library
import numpy as np
import scipy as sc
import pandas as pd
```

```python
train = pd.read_csv("train.csv", encoding="big5")
train
```

```python
filter1 = train["測項"]=="PM2.5"
pm25_train = train[filter1]
pm25_train.drop(['日期', '測站', '測項'], axis = 1, inplace=True)
pm25_train
```

```python
tempxlist=[]
tempylist=[]
for i in range(15):
    tempx=pm25_train.iloc[:,i:i+9]        #使用前9小时数据作为feature
    tempx.columns=np.array(range(9))
    tempy=pm25_train.iloc[:,i+9]         #使用第10个小数数据作为lable
    tempy.columns=['1']
    tempxlist.append(tempx)
    tempylist.append(tempy)
```

```python
xdata=pd.concat(tempxlist)     #feature数据
x=np.array(xdata,float)     
x  
```

```python
x.shape
```

```python
ydata=pd.concat(tempylist)      #lable数据
ydata.head()
y=(np.array(ydata,float))
y
```

```python
y.shape
```

```python
x = np.concatenate((np.ones((x.shape[0],1)),x), axis=1)      #在feature基础上加入bias
x
```

```python
# 初始化一个参数矩阵
w=np.zeros((len(x[0])))

#初始化一个learning rate
lr=0.1
iteration= 20000 #迭代10000次
s_grad=np.zeros(len(x[0]))
for i in range(iteration):
    tem=np.dot(x,w)     #&y^*&(预测值)
    loss=y-tem     
    grad=np.dot(x.transpose(),loss)*(-2)
    s_grad+=grad**2
    ada=np.sqrt(s_grad)
    w=w-lr*grad/ada
w    
```

```python
abs(loss).mean()
```

```python
w.shape
```

```python
pm25_train.to_csv("pm25_train.csv")
```

```python
test = pd.read_csv("test.csv", encoding="big5")
test
```

```python
filter2 = test["AMB_TEMP"]=="PM2.5"

pm25Test = test[filter2]

pm25Test.drop(['id_0', 'AMB_TEMP'],1,inplace=True)
pm25Test
```

```python
pm25Test.to_csv("pm25Test.csv")
```

```python
x_test=np.array(pm25Test,float)
x_test
```

```python
x_test_b=np.concatenate((np.ones((x_test.shape[0],1)),x_test),axis=1)
x_test
```

```python
y_star=np.dot(x_test_b,w)
y_star
```
